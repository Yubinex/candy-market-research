## Visualization

```python
# candies ordered by win percentage
sorted_winpercent = df.sort_values(by='Win Percent', ascending=False)
sorted_winpercent
```

###  Top 10 Candies

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=sorted_winpercent.head(10), x='Win Percent', y='Competitor Name', palette='viridis')
plt.xlabel('Win Percent')
plt.xlim(0, 100)
plt.ylabel('Candy')
plt.title('Candies ordered by Win Percent')
plt.show()
```

### Correlation Matrix

```python
# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

### Column Correlation with Win Percentage

```python
# Calculate correlation coefficients
correlation_matrix = df.corr(numeric_only=True)['Win Percent'].drop('Win Percent')

# Sort correlations in descending order
sorted_correlations = correlation_matrix.sort_values(ascending=False)

# Color palette for the bars
colors = sns.color_palette("seismic", len(sorted_correlations))

# Bar chart with colored bars
plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_correlations.index, y=sorted_correlations.values, palette=colors)
plt.title("Correlation of Columns with Win Percentage")
plt.xlabel("Columns")
plt.ylabel("Correlation with Win Percentage")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Top Correlation Pairs between Columns and Win Percentage

```python
from tabulate import tabulate

# Calculate correlation coefficients
correlation_matrix = df.corr(numeric_only=True)

# Exclude self-correlations and get the top 10 correlations
top_correlations = (correlation_matrix
                    .unstack()
                    .sort_values(ascending=False)
                    .drop_duplicates()
                    .iloc[1:11])  # Exclude self-correlation

# Organize correlations into a list of lists for tabulate
correlation_list = []
for (col1, col2), correlation in top_correlations.items():
    correlation_list.append([f"{col1} - {col2}", f"{correlation:.3f}"])

# Print the top 10 correlations using tabulate
print("Top 10 Correlations:")
print(tabulate(correlation_list, headers=["Columns", "Correlation"], tablefmt="pretty"))
```

### Column Percentages for Win Percent 

```python
# Convert 'Win Percent' column to numeric
df['Win Percent'] = pd.to_numeric(df['Win Percent'], errors='coerce')

# Calculate the total number of rows
total_rows = len(df)

# Exclude 'Win Percent' column from numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns
numeric_columns = numeric_columns[numeric_columns != 'Win Percent']

# Calculate percentages for columns when Win Percent > 50%
above_50 = df[df['Win Percent'] > 50]
above_50_percentages = above_50[numeric_columns].sum() / total_rows * 100

# Calculate percentages for columns when Win Percent <= 50%
below_50 = df[df['Win Percent'] <= 50]
below_50_percentages = below_50[numeric_columns].sum() / total_rows * 100

# Print the percentages
print("Percentages for columns when Win Percent > 50%:")
print(above_50_percentages)

print("\nPercentages for columns when Win Percent <= 50%:")
print(below_50_percentages)
```

```python
# Create a bar plot for above 50% percentages
plt.figure(figsize=(10, 5))
sns.barplot(x=above_50_percentages.index, y=above_50_percentages.values)
plt.title("Percentages for Columns when Win Percent > 50%")
plt.xticks(rotation=45)
plt.ylabel("Percentage")
plt.tight_layout()
plt.show()

# Create a bar plot for below 50% percentages
plt.figure(figsize=(10, 5))
sns.barplot(x=below_50_percentages.index, y=below_50_percentages.values)
plt.title("Percentages for Columns when Win Percent <= 50%")
plt.xticks(rotation=45)
plt.ylabel("Percentage")
plt.tight_layout()
plt.show()
```
